```python
# dependencies and data setup
import pandas as pd

# File to Load (Remember to change the path if needed.)
school_data_to_load = "resources/schools_complete.csv"
student_data_to_load = "resources/students_complete.csv"

# Read the School Data and Student Data and store into a Pandas DataFrame
school_df = pd.read_csv(school_data_to_load)
student_df = pd.read_csv(student_data_to_load)
```

```python
# clean student names

# list of prefixes and suffixes to remove
prefixes_suffixes = ["Dr. ", "Mr. ","Ms. ", "Mrs. ", "Miss ", " MD", " DDS", " DVM", " PhD"]

# replace prefixes/suffixes with an empty space
for word in prefixes_suffixes:
    student_df["student_name"] = student_df["student_name"].str.replace(word,"")

student_df.head()
```

```python
### Replace the 9th grade reading and math scores at Thomas High School with NaN.
# Install numpy using conda install numpy or pip install numpy. 

# Step 1. Import numpy as np.
import numpy as np
```

```python
# Step 2. Use the loc method on the student_df to select all the reading scores from the 9th grade at Thomas High School and replace them with NaN.

student_df.loc[(student_df["grade"]  == "9th") & (student_df["school_name"] == "Thomas High School"), "reading_score"] = np.nan
```

```python
#  Step 3. Refactor the code in Step 2 to replace the math scores with NaN.

student_df.loc[(student_df["grade"]  == "9th") & (student_df["school_name"] == "Thomas High School"), "math_score"] = np.nan
    
```

## Deliverable 2 : Repeat the school district analysis

### District Summary

```python
# Combine school & student data into a single dataset
full_df = pd.merge(student_df, school_df, how="left", on=["school_name", "school_name"])
full_df.head()
```

```python
# Calculate the Totals (Schools and Students)
school_count = len(full_df["school_name"].unique())
student_count = full_df["Student ID"].count()

# Calculate the Total Budget
total_budget = school_df["budget"].sum()

# Calculate the Average Scores using the "clean_student_data".
average_reading_score = full_df["reading_score"].mean()
average_math_score = full_df["math_score"].mean()
```

```python
# Step 1. Get the number of students that are in ninth grade at Thomas High School.

# These students have no grades. 
nogrades_count = full_df.loc[(full_df["grade"]  == "9th") & (full_df["school_name"] == "Thomas High School")]["student_name"].count()

# Get the total student count 
student_count = full_df["student_name"].count()
```

```python
# Step 2. Subtract the number of students that are in ninth grade at 
# Thomas High School from the total student count to get the new total student count.

new_student_count = student_count - nogrades_count
```

```python
# Calculate the passing rates using the "clean_student_data".

passing_math_count = full_df[(full_df["math_score"] >= 70)].count()["student_name"]
passing_reading_count = full_df[(full_df["reading_score"] >= 70)].count()["student_name"]
```

```python
# Step 3. Calculate the passing percentages with the new total student count.

# Calculate the students who passed both reading and math.
passing_math_reading = full_df[(full_df["math_score"] >= 70)
                                               & (full_df["reading_score"] >= 70)]

# Calculate the number of students that passed both reading and math.
overall_passing_math_reading_count = passing_math_reading["student_name"].count()
```

```python
# Step 4. Calculate the overall passing percentage with new total student count.

passing_math_percentage = passing_math_count / new_student_count * 100
passing_reading_percentage = passing_reading_count / new_student_count * 100
overall_passing_percentage = overall_passing_math_reading_count / new_student_count * 100

# Create a DataFrame
district_summary_df = pd.DataFrame(
          [{"Total Schools": school_count, 
          "Total Students": student_count, 
          "Total Budget": total_budget,
          "Average Math Score": average_math_score, 
          "Average Reading Score": average_reading_score,
          "% Passing Math": passing_math_percentage,
         "% Passing Reading": passing_reading_percentage,
        "% Overall Passing": overall_passing_percentage}])
```

```python
# Format the "Total Students" to have the comma for a thousands separator.
district_summary_df["Total Students"] = district_summary_df["Total Students"].map("{:,}".format)

# Format the "Total Budget" to have the comma for a thousands separator, a decimal separator and a "$".
district_summary_df["Total Budget"] = district_summary_df["Total Budget"].map("${:,.2f}".format)

# Format the columns.
district_summary_df["Average Math Score"] = district_summary_df["Average Math Score"].map("{:.1f}".format)
district_summary_df["Average Reading Score"] = district_summary_df["Average Reading Score"].map("{:.1f}".format)
district_summary_df["% Passing Math"] = district_summary_df["% Passing Math"].map("{:.1f}".format)
district_summary_df["% Passing Reading"] = district_summary_df["% Passing Reading"].map("{:.1f}".format)
district_summary_df["% Overall Passing"] = district_summary_df["% Overall Passing"].map("{:.1f}".format)

# Display the data frame
district_summary_df
```

##  School Summary

```python
# Determine the School Type
per_school_types = school_df.set_index(["school_name"])["type"]

# Calculate the total student count.
per_school_counts = full_df["school_name"].value_counts()

# Calculate the total school budget and per capita spending
per_school_budget = full_df.groupby(["school_name"]).mean()["budget"]

# Calculate the per capita spending.
per_school_capita = per_school_budget / per_school_counts

# Calculate the average test scores.
per_school_math = full_df.groupby(["school_name"]).mean()["math_score"]
per_school_reading = full_df.groupby(["school_name"]).mean()["reading_score"]

# Calculate the passing scores by creating a filtered DataFrame.
per_school_passing_math = full_df[(full_df["math_score"] >= 70)]
per_school_passing_reading = full_df[(full_df["reading_score"] >= 70)]

# Calculate the number of students passing math and passing reading by school.
per_school_passing_math = per_school_passing_math.groupby(["school_name"]).count()["student_name"]
per_school_passing_reading = per_school_passing_reading.groupby(["school_name"]).count()["student_name"]

# Calculate the percentage of passing math and reading scores per school.
per_school_passing_math = per_school_passing_math / per_school_counts * 100
per_school_passing_reading = per_school_passing_reading / per_school_counts * 100

# Calculate the students who passed both reading and math.
per_passing_math_reading = full_df[(full_df["reading_score"] >= 70)
                                               & (full_df["math_score"] >= 70)]

# Calculate the number of students passing math and passing reading by school.
per_passing_math_reading = per_passing_math_reading.groupby(["school_name"]).count()["student_name"]

# Calculate the percentage of passing math and reading scores per school.
per_overall_passing_percentage = per_passing_math_reading / per_school_counts * 100
# Create the DataFrame
per_school_summary_df = pd.DataFrame({
    "School Type": per_school_types,
    "Total Students": per_school_counts,
    "Total School Budget": per_school_budget,
    "Per Student Budget": per_school_capita,
    "Average Math Score": per_school_math,
    "Average Reading Score": per_school_reading,
    "% Passing Math": per_school_passing_math,
    "% Passing Reading": per_school_passing_reading,
    "% Overall Passing": per_overall_passing_percentage})


# per_school_summary_df.head()
# Format the Total School Budget and the Per Student Budget
per_school_summary_df["Total School Budget"] = per_school_summary_df["Total School Budget"].map("${:,.2f}".format)
per_school_summary_df["Per Student Budget"] = per_school_summary_df["Per Student Budget"].map("${:,.2f}".format)

# Display the data frame
per_school_summary_df
```

```python
# Step 5.  Get the number of 10th-12th graders from Thomas High School (THS).

THS_new_df = full_df.loc[(full_df["school_name"] == "Thomas High School") & (full_df["grade"] != "9th")]

THS_new_student_count = THS_new_df["student_name"].count()
```

```python
# Step 6. Get all the students passing math from THS

THS_passed_math_count = THS_new_df.loc[THS_new_df["math_score"] >= 70]["student_name"].count()
```

```python
# Step 7. Get all the students passing reading from THS

THS_passed_reading_count = THS_new_df.loc[THS_new_df["reading_score"] >= 70]["student_name"].count()
```

```python
# Step 8. Get all the students passing math and reading from THS

THS_passed_both_count = THS_new_df.loc[(THS_new_df["math_score"] >= 70) & (THS_new_df["reading_score"] >= 70)]["student_name"].count()
```

```python
# Step 9. Calculate the percentage of 10th-12th grade students passing math from Thomas High School. 

THS_passed_math_pct = THS_passed_math_count / THS_new_student_count * 100
```

```python
# Step 10. Calculate the percentage of 10th-12th grade students passing reading from Thomas High School.

THS_passed_reading_pct = THS_passed_reading_count / THS_new_student_count * 100
```

```python
# Step 11. Calculate the overall passing percentage of 10th-12th grade from Thomas High School. 

THS_passed_both_pct = THS_passed_both_count / THS_new_student_count * 100
```

```python
# Step 12. Replace the passing math percent for Thomas High School in the per_school_summary_df.

per_school_summary_df.loc["Thomas High School" , "% Passing Math"] = THS_passed_math_pct
```

```python
# Step 13. Replace the passing reading percentage for Thomas High School in the per_school_summary_df.

per_school_summary_df.loc["Thomas High School" , "% Passing Reading"] = THS_passed_reading_pct
```

```python
# Step 14. Replace the overall passing percentage for Thomas High School in the per_school_summary_df.

per_school_summary_df.loc["Thomas High School" , "% Overall Passing"] = THS_passed_both_pct

# per_school_summary_df
per_school_summary_df.tail()
```

```python
# Group each school Series by the school name for the average reading score.

grade_9_reading = grade_9.groupby(["school_name"]).mean()["reading_score"]
grade_10_reading = grade_10.groupby(["school_name"]).mean()["reading_score"]
grade_11_reading = grade_11.groupby(["school_name"]).mean()["reading_score"]
grade_12_reading = grade_12.groupby(["school_name"]).mean()["reading_score"]
```

```python
# Combine each Series for average math scores by school into single data frame.

math_scores_by_grade = pd.DataFrame({
               "9th": grade_9_math,
               "10th": grade_10_math,
               "11th": grade_11_math,
               "12th": grade_12_math})
```

```python
# Combine each Series for average reading scores by school into single data frame.

reading_scores_by_grade = pd.DataFrame({
               "9th": grade_9_reading,
               "10th": grade_10_reading,
               "11th": grade_11_reading,
               "12th": grade_12_reading})
```

```python
# Format each grade column.
math_scores_by_grade["9th"] = math_scores_by_grade["9th"].map("{:.1f}".format)
math_scores_by_grade["10th"] = math_scores_by_grade["10th"].map("{:.1f}".format)
math_scores_by_grade["11th"] = math_scores_by_grade["11th"].map("{:.1f}".format)
math_scores_by_grade["12th"] = math_scores_by_grade["12th"].map("{:.1f}".format)

reading_scores_by_grade["9th"] = reading_scores_by_grade["9th"].map("{:.1f}".format)
reading_scores_by_grade["10th"] = reading_scores_by_grade["10th"].map("{:.1f}".format)
reading_scores_by_grade["11th"] = reading_scores_by_grade["11th"].map("{:.1f}".format)
reading_scores_by_grade["12th"] = reading_scores_by_grade["12th"].map("{:.1f}".format)

# Remove the indexes
math_scores_by_grade.index.name = None
reading_scores_by_grade.index.name = None
```

## Scores by School Spending

```python
# Establish the spending bins and group names.

spending_bins = [0, 585, 630, 645, 675]
group_names = ["<$586", "$586-630", "$631-645", "$646-675"]
```

```python
# Categorize spending based on the bins.

per_school_summary_df["Spending Range Per Student"] = pd.cut(per_school_capita, spending_bins, labels=group_names)
```

```python
# Calculate average scores based on per-student funding

bin_math_scores = per_school_summary_df.groupby(["Spending Range Per Student"]).mean()["Average Math Score"]

bin_reading_scores = per_school_summary_df.groupby(["Spending Range Per Student"]).mean()["Average Reading Score"]

bin_math_passed = per_school_summary_df.groupby(["Spending Range Per Student"]).mean()["% Passing Math"]

bin_reading_passed = per_school_summary_df.groupby(["Spending Range Per Student"]).mean()["% Passing Reading"]

bin_both_passed = per_school_summary_df.groupby(["Spending Range Per Student"]).mean()["% Overall Passing"]
```

```python
# Create dataframe of scores based on per-student funding

spending_summary_df = pd.DataFrame({
          "Average Math Score" : bin_math_scores,
          "Average Reading Score": bin_reading_scores,
          "% Passing Math": bin_math_passed,
          "% Passing Reading": bin_reading_passed,
          "% Overall Passing": bin_both_passed})
```

```python
# Format the DataFrame 

spending_summary_df["Average Math Score"] = spending_summary_df["Average Math Score"].map("{:.1f}".format)

spending_summary_df["Average Reading Score"] = spending_summary_df["Average Reading Score"].map("{:.1f}".format)

spending_summary_df["% Passing Math"] = spending_summary_df["% Passing Math"].map("{:.0f}".format)

spending_summary_df["% Passing Reading"] = spending_summary_df["% Passing Reading"].map("{:.0f}".format)

spending_summary_df["% Overall Passing"] = spending_summary_df["% Overall Passing"].map("{:.0f}".format)

spending_summary_df
```
