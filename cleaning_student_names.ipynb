```python
import pandas as pd
```

```python
# load in data as dataframe
student_df = pd.read_csv("resources/students_complete.csv")
student_df.head(20)
```

```python
# save student names to a list
names = student_df["student_name"].tolist()
names[:5]
```

```python
# create list of names that need fixing
names_to_fix = []

# add names to list if they are too long
for name in names:
    
    if len(name.split()) >= 3:
        names_to_fix.append(name)

# number of names that need fixing
len(names_to_fix)
```

```python
# percentage of names that need fixing
len(names_to_fix)/len(names)
```

```python
names_to_fix[:10]
```

```python
# create a new list of prefixes shorter than or equal to 4 letters
prefixes = []

for name in names_to_fix:
    if len(name.split()[0]) <= 4:
        prefixes.append(name.split()[0])

print(prefixes[:50])
```

```python
# new list of suffixes less than or equal to 3 letters
suffixes = []
for name in names_to_fix:
    if len(name.split()[-1]) <= 3:
        suffixes.append(name.split()[-1])

print(suffixes[:50])
```

```python
# get all unique prefixes
print(set(prefixes))
```

```python
# number of unique prefixes
len(set(prefixes))
```

```python
# get all unique suffixes
print(set(suffixes))
```

```python
# number of unique suffixes
len(set(suffixes))
```

```python
# list of all prefixes and suffixes to remove
presuf = ["Dr. ", "Mr. ","Ms. ", "Mrs. ", "Miss ", " MD", " DDS", " DVM", " PhD"]
```

```python
# replace pre/suffixes with an empty space
for word in presuf:
    student_df["student_name"] = student_df["student_name"].str.replace(word,"")
```

```python
student_df.head(20)
```

```python
# new list of cleaned names
check_names = student_df["student_name"].tolist()
print(check_names[:50])
```

```python
# new list for names that still have prefixes/suffixes
doublecheck_names = []

# add names >= 3 letters to the list
for name in check_names:
    if len(name.split()) >= 3:
        doublecheck_names.append(name)
        
# number of fixed names
len(doublecheck_names)
```
